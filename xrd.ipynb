```python
#importation du cif et des packages n√©cessaires

from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.analysis.diffraction import xrd
from matplotlib import pyplot as plt

with MPRester("im80MbCif8QDmhc2y") as m:
    structure = m.get_structure_by_material_id("mp-38011")
    w = CifWriter(structure)
    w.write_file('KHS_mp-38011_symmetrized')
   
```

```python
# Obtention des caract√©ristiques de la diffraction 
dif = xrd.XRDCalculator(wavelength=1.54060)
caractDif = dif.get_pattern(structure)

# aper√ßu du diffractogramme ()
dif.show_plot(structure,annotate_peaks=False)
```

# Caract√©ristiques du premier pic 

```python
print("Angle : " + str(caractDif.x[0]) + "¬∞")
print("Intensite : " + str(caractDif.y[0]))
print("Indices hkl : " + str(caractDif.hkls[0]))
```
